# Preprocessing the data

## Missing values imputation



```python
# lets check the rows that having missing valuse 

df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
# lets check the rows that having high percentage of missing valuse 

round(df.isnull().sum()/len(df)*100,2)
```

```python
# drop nan values from 2 columns that having high percentage of missing values

df.dropna(subset=['gross','budget'],inplace=True)
df.isnull().sum()
```

```python
df = df[df.isnull().sum(axis=1) <=2]
df.isnull().sum()
```

```python
# lets impute the missing values 

#using mean for numerical columns
df['num_critic_for_reviews'].fillna(df['num_critic_for_reviews'].mean(),inplace=True)
df['duration'].fillna(df['duration'].mean(), inplace=True)

# using mode for categorical columns
df['language'].fillna(df['language'].mode()[0], inplace=True)

# As we know that We cannot use statistical values for imputing the missing values of actor names, so we will replace the 
# actor names with "Unknown Actor"

df['actor_2_name'].fillna('Unknown Actor', inplace=True)
df['actor_3_name'].fillna('Unknown Actor', inplace=True)

```

# Feature Engineering

```python
# Lets convert the gross and budget from $ to Million $ to make our analysis easier
df['budget'] = df['budget']/1000000
df['gross'] = df['gross']/1000000
```

```python
# Create new feature(profit) using budget and gross

df['profit'] = df['gross'] - df['budget']
```

```python
df[['profit' ,'country']].sort_values(by='profit', ascending = False).head(10)
```

```python
df[['profit','movie_title']].sort_values(by='profit',ascending=False).head(10)
```

```python
# Remove duplicated data

print('No. of rows befor removing duplicates: ', df.shape[0])

df.drop_duplicates(subset=None , keep='first', inplace=True)

print('No. of rows after removing duplicates', df.shape[0])
```
